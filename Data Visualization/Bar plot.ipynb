```python
import matplotlib.pyplot as plt
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
```

```python
import numpy as np

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years))
width = 0.25

plt.bar(x-width, sachin, label="Sachine", width=width )
plt.bar(x, sehwag, label="Sehwag", width=width )
plt.bar(x+width, kohli, label="Virat", width=width )

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")

plt.xticks(x,years)
plt.legend()
```

```python
# Assignment   
import numpy as np

years = [2001, 2005, 2010, 2015, 2019]

sachin_runs = [1200, 900, 1100, 800, 700]
kohli_runs = [0, 500, 1000, 1200, 1300]
sehwag_runs = [1100, 1300, 900, 400, 200]

x = np.arange(len(years))
w =0.25
plt.bar(x-w , sehwag_runs, label="Sehwag" ,width=w , color="#bbd020")
plt.bar(x , sachin_runs, label="Sachine" ,width=w , color="#1dc1e1")
plt.bar(x+w , kohli_runs, label="Virat" ,width=w , color="#a01de1")

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")

plt.tight_layout()
plt.legend()
plt.xticks(x,years)
```
