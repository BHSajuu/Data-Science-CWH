```python
def shuffle_and_split(data, test_ratio):
    np.random.seed(42) # Set the seed for reproducibility 
    shuffled_idxs = np.random.permutation(len(data))  # this return the shuffle indexes
    test_set_size = int(len(data) * test_ratio)
    test_idxs = shuffled_idxs[ : test_set_size]
    train_idxs = shuffled_idxs[test_set_size : ]
    return data.iloc[train_idxs], data.iloc[test_idxs]
```

```python
df['income_cat'] = pd.cut(df['median_income'], bins=[0, 1.5, 3.0, 4.5, 6.0, np.inf], labels = [1, 2, 3, 4, 5])
```

```python
df.head()
```

```python
# visualize  the incaome_cat in matplotlib
import matplotlib.pyplot as plt

df["income_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.title("Income Categories Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number of Instances")
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```
